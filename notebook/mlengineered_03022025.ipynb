```python
df['parsed_os'] = df['Device Information'].apply(lambda x: parse_ua(x).os)
```

```python
df.head()
```

```python
def os_version(row):
    
    ua = row['Device Information']
    parsed_ua = parse_ua(ua)
    os_name = parsed_ua.os['name']
    os_version = parsed_ua.os['version']
    version = os_version.split(".")[0]
  
    
    return f"{os_name} {version}"



df['version_os'] = df.apply(os_version,axis = 1)
    
```

```python
def browser_version(row):
    
    ua = row['Device Information']
    parsed_ua = parse_ua(ua)
    browser_name = parsed_ua.browser['name']
    browser_version = parsed_ua.browser['version']
    version = browser_version.split(".")[0]
    
    return f"{browser_name} {version}"

df['browser_os'] = df.apply(browser_version,axis = 1)
    
```

```python
def device_version(row):
    
    ua = row['Device Information']
    parsed_ua = parse_ua(ua)
    device_type = parsed_ua.device['type']
    if device_type == None:
        device_type = "computer"
    return device_type

df['device_type'] = df.apply(device_version,axis = 1)
    
    
```

```python
import pandas as pd
import numpy as np

# 1. Extract IP Prefixes (for both source and destination)
def extract_ip_prefix(ip_address): 
        part1 = ip_address.split(".")[0]
        part2 = ip_address.split(".")[1]
        part3 = f"{part1}{part2}"
        return part3

df['Source IP Prefix'] = df['Source IP Address'].apply(lambda x: extract_ip_prefix(x))
df['Destination IP Prefix'] = df['Destination IP Address'].apply(lambda x: extract_ip_prefix(x))
```

```python
# 2. Create Vocabulary (unique prefixes)
source_prefixes = df['Source IP Prefix'].dropna().unique()
dest_prefixes = df['Destination IP Prefix'].dropna().unique()
all_prefixes = np.concatenate([source_prefixes, dest_prefixes]) # Combine and get unique values
unique_prefixes = np.unique(all_prefixes)

prefix_to_index = {prefix: index for index, prefix in enumerate(unique_prefixes)} # Mapping for lookup
```

```python
# 3. Convert Prefixes to Indices
df['Source Prefix Index'] = df['Source IP Prefix'].apply(lambda x: prefix_to_index.get(x) if pd.notnull(x) else -1) # -1 for missing values
df['Dest Prefix Index'] = df['Destination IP Prefix'].apply(lambda x: prefix_to_index.get(x) if pd.notnull(x) else -1)
```

```python

# 4. TensorFlow Embedding (Example)
embedding_dim = 32  # Adjust embedding dimension
```

```python
import tensorflow as tf 
```

```python
# Create embedding layer
embedding_layer = tf.keras.layers.Embedding(len(unique_prefixes), embedding_dim)
```

```python
# Function to get embeddings (for use in a tf.data.Dataset pipeline or inside the model)
def get_embeddings(source_index, dest_index):
    source_embedding = embedding_layer(source_index)
    dest_embedding = embedding_layer(dest_index)
    return source_embedding, dest_embedding
```

```python

well_known_ports = {
    20: 'FTP (File Transfer Protocol) - Transfert de fichiers',
    21: 'FTP (File Transfer Protocol) - Connexion de contrôle',
    22: 'SSH (Secure Shell) - Connexion sécurisée',
    23: 'Telnet - Connexion à distance non sécurisée',
    25: 'SMTP (Simple Mail Transfer Protocol) - Envoi d\'emails',
    53: 'DNS (Domain Name System) - Résolution de noms de domaine',
    80: 'HTTP (HyperText Transfer Protocol) - Trafic web non sécurisé',
    110: 'POP3 (Post Office Protocol) - Récupération d\'emails',
    143: 'IMAP (Internet Message Access Protocol) - Récupération d\'emails',
    443: 'HTTPS (HTTP Secure) - Trafic web sécurisé',
    3389: 'RDP (Remote Desktop Protocol) - Connexion bureau à distance'}
 

registered_ports = {
    1024: 'Service réservé',
    1080: 'SOCKS Proxy',
    1433: 'Microsoft SQL Server',
    1434: 'Microsoft SQL Monitor',
    1521: 'Oracle Database Listener',
    1723: 'Point-to-Point Tunneling Protocol (PPTP)',
    2049: 'Network File System (NFS)',
    2082: 'cPanel (interface d\'hébergement web)',
    2083: 'cPanel sécurisé (HTTPS)',
    2181: 'Apache ZooKeeper',
    3306: 'MySQL Database Service',
    5432: 'PostgreSQL Database Service',
    5900: 'Virtual Network Computing (VNC)',
    5984: 'CouchDB Database Service',
    6379: 'Redis Key-Value Store',
    8080: 'HTTP alternatif (HTTP Proxies)',
    8443: 'HTTPS alternatif (HTTP Proxies sécurisés)',
    8888: 'HTTP alternatif'}
 

def categorize_port(port):
    if port in well_known_ports:
        return f"Bien connu : {well_known_ports[port]}"
    elif port in registered_ports:
        return f"Enregistré : {registered_ports[port]}"
    elif 0 <= port <= 1023:
        return 'Bien connu : Autre'
    elif 1024 <= port <= 49151:
        return 'Enregistré : Autre'
    elif 49152 <= port <= 65535:
        return 'Dynamique/Privé'
    else:
        return 'Inconnu'
 

df['Port Category'] = df['Destination Port'].apply(categorize_port)
 

df.head()
```
